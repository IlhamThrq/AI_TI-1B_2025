### Latihan 2: Permen

Mario memiliki sekantong permen. Terdapat lima jenis permen, masing-masing diberi nama di bawah ini. Buat diagram yang menunjukkan persentase peluang bahwa kita akan mengeluarkan permen Snickers dari kantong jika mario melakukan pengambilan acak. Sebutkan peluang memilih permen Snickers.

```python
import matplotlib.pyplot as plt

# Data
candy_names = ['Kit Kat', 'Snickers', 'Milky Way', 'Toblerone', 'Twix']
candy_counts = [52, 39, 78, 13, 78]

# Hitung total dan persentase untuk Snickers
total = sum(candy_counts)
snickers_index = candy_names.index("Snickers")
snickers_percentage = (candy_counts[snickers_index] / total) * 100

# Pie chart
plt.figure(figsize=(8, 8))
colors = ['#e74c3c', '#3498db', '#2ecc71', '#f1c40f', '#9b59b6']
explode = [0, 0.1, 0, 0, 0]  # Menonjolkan Snickers

plt.pie(candy_counts, labels=candy_names, autopct='%1.1f%%',
        startangle=140, colors=colors, explode=explode,
        shadow=True)

plt.title('Peluang Mengambil Permen Snickers dari Kantong', fontsize=14)
plt.tight_layout()
plt.show()

# Tampilkan peluang memilih Snickers
print(f"Peluang memilih Snickers: {snickers_percentage:.2f}%")
```

### Latihan 3: Makanan

Restoran memiliki menu makanan penutup yang terlalu besar. Mereka ingin memotong beberapa item dari menu. Untuk membuat sebagian besar pelanggan mereka senang, mereka ingin menghapus hanya tiga makanan penutup yang paling tidak populer dari menu. Kita memiliki daftar makanan penutup yang disajikan restoran, serta hitungan berapa kali makanan penutup tersebut dijual dalam seminggu terakhir. Buat visualisasi yang menunjukkan popularitas relatif dari makanan penutup. Sebutkan tiga makanan penutup yang harus disingkirkan.

```python
import matplotlib.pyplot as plt

# Data
dessert_sales = {
    'Lava Cake': 14,
    'Mousse': 5,
    'Chocolate Cake': 12,
    'Ice Cream': 19,
    'Truffles': 6,
    'Brownie': 8,
    'Chocolate Chip Cookie': 12,
    'Chocolate Pudding': 9,
    'Souffle': 10,
    'Chocolate Cheesecake': 17,
    'Chocolate Chips': 2,
    'Fudge': 9,
    'Mochi': 13,
}

# Sort data by sales (ascending)
sorted_desserts = sorted(dessert_sales.items(), key=lambda x: x[1])
dessert_names = [item[0] for item in sorted_desserts]
dessert_counts = [item[1] for item in sorted_desserts]

# Tandai 3 yang paling tidak populer
colors = ['brown' if i < 3 else 'orange' for i in range(len(dessert_names))]

# Bar chart
plt.figure(figsize=(12, 6))
bars = plt.barh(dessert_names, dessert_counts, color=colors)
plt.xlabel('Jumlah Penjualan')
plt.title('Popularitas Makanan Penutup Minggu Ini')

# Tambahkan anotasi pada batang
for bar in bars:
    width = bar.get_width()
    plt.text(width + 0.5, bar.get_y() + bar.get_height() / 2,
             str(int(width)), va='center')

plt.tight_layout()
plt.gca().invert_yaxis()  # Terbalikkan urutan biar yang terbawah paling populer
plt.show()

# Tampilkan 3 makanan penutup yang harus disingkirkan
to_remove = dessert_names[:3]
print("Tiga makanan penutup yang harus disingkirkan:", to_remove)
```

### Latihan 4: Penggunaan CPU

Kita memiliki penggunaan CPU rata-rata per jam untuk komputer pekerja selama seminggu. Setiap baris data mewakili satu hari dalam seminggu yang dimulai dengan Senin. Setiap kolom data adalah satu jam dalam sehari dimulai dengan 0 menjadi tengah malam.
Buat bagan yang menunjukkan penggunaan CPU selama seminggu. Anda harus dapat menjawab pertanyaan-pertanyaan berikut menggunakan bagan:
•	Jam berapa pekerja biasanya makan siang?
•	Apakah pekerja tersebut bekerja pada akhir pekan?
•	Pada hari apa pekerja mulai bekerja pada komputer mereka pada malam hari?


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Data penggunaan CPU
cpu_usage = [
    [2, 2, 4, 2, 4, 1, 1, 4, 4, 12, 22, 23, 45, 9,
        33, 56, 23, 40, 21, 6, 6, 2, 2, 3],  # Monday
    [1, 2, 3, 2, 3, 2, 3, 2, 7, 22, 45, 44, 33, 9,
        23, 19, 33, 56, 12, 2, 3, 1, 2, 2],  # Tuesday
    [2, 3, 1, 2, 4, 4, 2, 2, 1, 2,  5, 31, 54, 7, 6,
        34, 68, 34, 49, 6, 6, 2, 2, 3],  # Wednesday
    [1, 2, 3, 2, 4, 1, 2, 4, 1, 17, 24, 18, 41, 3, 44,
        42, 12, 36, 41, 2, 2, 4, 2, 4],  # Thursday
    [4, 1, 2, 2, 3, 2, 5, 1, 2, 12, 33, 27, 43, 8,
        38, 53, 29, 45, 39, 3, 1, 1, 3, 4],  # Friday
    [2, 3, 1, 2, 2, 5, 2, 8, 4, 2, 3, 1, 5, 1, 2, 3,
        2, 6, 1, 2, 2, 1, 4, 3],         # Saturday
    [1, 2, 3, 1, 1, 3, 4, 2, 3, 1, 2, 2, 5, 3, 2, 1,
        4, 2, 45, 26, 33, 2, 2, 1],       # Sunday
]

# Label
days = ['Monday', 'Tuesday', 'Wednesday',
        'Thursday', 'Friday', 'Saturday', 'Sunday']
hours = list(range(24))

# Plot heatmap
plt.figure(figsize=(16, 6))
sns.heatmap(cpu_usage, cmap='YlOrRd', xticklabels=hours, yticklabels=days,
            linewidths=0.3, linecolor='gray', annot=True, fmt='d')
plt.title('Penggunaan CPU per Jam Selama Seminggu', fontsize=16)
plt.xlabel('Jam (0 = Tengah Malam)', fontsize=12)
plt.ylabel('Hari', fontsize=12)
plt.tight_layout()
plt.show()
```
